```python
inVar=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']

from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) 
    for i in range(X.shape[1])]
    return(vif)
```

```python
# For checking Multi-Collinearity

X =df[['Glucose','Insulin','SkinThickness','Pregnancies','BloodPressure','BMI','DiabetesPedigreeFunction','Age']]
calc_vif(X)
```

```python
X =df[['Pregnancies','Glucose','SkinThickness','Insulin','DiabetesPedigreeFunction','Age']]
calc_vif(X)
```

```python
X =df[['Glucose','Pregnancies','Insulin','Age','DiabetesPedigreeFunction']]
calc_vif(X)
```

```python
x= df[['Pregnancies','Glucose','SkinThickness','Insulin','DiabetesPedigreeFunction','Age']]
y = df['Diabetes Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=101)
from sklearn.linear_model import LogisticRegression
logm1 = LogisticRegression()
logm1.fit(x_train,y_train)
```

```python
y_pre1 = logm1.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pre1))
```
